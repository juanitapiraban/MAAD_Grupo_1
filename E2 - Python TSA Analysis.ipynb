```python
# Importa Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import csv
import sys
import plotly as pt
import plotly.express as px

%matplotlib inline
sns.set()    
```

```python
# Leer el archivo 'datos.csv'
df = pd.read_csv("Vicente_Fernandez_TSA.csv")

# Ajuste de Dataframe
df.columns  =["Mes", "Vicente"]


df.Mes = pd.to_datetime(df.Mes)
df.set_index('Mes', inplace=False)

# Informaci√≥n del dataframe
df.info()
```

```python
df.plot(figsize=(17,4),linewidth=3, fontsize=15)
```

```python
#Identificaci√≥n de tendencias en series temporales - Promedio Movil -12 meses
df.rolling(12).mean().plot(figsize=(17,5), linewidth=5, fontsize=20)
plt.xlabel('Mes', fontsize=15);
```

```python
df.diff().plot(figsize=(10,5), linewidth=5, fontsize=20)
plt.xlabel('Mes', fontsize=20);
```
